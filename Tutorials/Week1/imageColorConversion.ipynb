```python
# Load color image
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2

myImg = cv2.imread('example_image.jpg')
```

```python
# Display original image
plt.figure()
plt.imshow(myImg)
plt.axis('off')
plt.title('Original Image')
```

```python
# Resize image to quarter of original size and display
quartImg = cv2.resize(myImg, None, fx=0.25, fy=0.25)
plt.figure()
plt.imshow(quartImg)
plt.axis('off')
plt.title('Quarter Size Image')
```

```python
# Seperate and display Red, Green and Blue channels
r, g, b = cv2.split(myImg)
print(r.shape, g.shape, b.shape)
```

```python
# Convert img to grayscale and display
grayImg = cv2.cvtColor(myImg, cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(grayImg, cmap='gray')
plt.axis('off')
plt.title('Grayscale Image')
```
